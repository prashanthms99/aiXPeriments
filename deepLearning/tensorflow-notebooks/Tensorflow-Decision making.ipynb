## Tesnsorflow Decision making 



```
import tensorflow as tf

print(tf.__version__)
```

###tf.cond

tf.cond(
    pred,
    true_fn=None,
    false_fn=None,
    strict=False,
    name=None,
    fn1=None,
    fn2=None
)

Return **true_fn()** if the predicate pred is true else **false_fn()**. (deprecated arguments)

```
a = tf.constant(3)
b = tf.constant(4)
z = tf.multiply(a, b)
result = tf.cond(a < b, lambda: tf.add(a, z), lambda: tf.square(b))
print(result)

with tf.Session() as sess:
   res=sess.run(result)
print(res)
```

###tf.less

tf.less(
    x,
    y,
    name=None
)

Returns the truth value of (x < y) element-wise.

```
x = tf.constant(9)
y = tf.constant(8)

r = tf.less(x,y)

with tf.Session() as sess:
  res=sess.run(r)
  
print(res)
```

###tf.less_equal

tf.less_equal(
    x,
    y,
    name=None
)

Returns the truth value of (x <= y) element-wise.

```
x = tf.constant(9)
y = tf.constant(9)

r = tf.less_equal(x,y)

with tf.Session() as sess:
  res=sess.run(r)
  
print(res)
```

###tf.case
tf.case( pred_fn_pairs, default=None,exclusive=False, strict=False, name='case')

Create a case operation.

The **pred_fn_pairs** parameter is a dict or list of pairs of size **N**.

```
x = tf.constant(17)
y = tf.constant(23)
f1 = lambda: tf.constant(17)
f2 = lambda: tf.constant(23)
r = tf.case([(tf.less(x, y), f1)], default=f2)
print(r)
```

```
with tf.Session() as sess:
  result=sess.run(r)
print(result)
```
